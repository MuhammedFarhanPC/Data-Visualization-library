# **Line Chart with Annotation**

```python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# Create a line chart
plt.figure(figsize=(8, 6)) #Creates a new figure with a size of 8 inches by 6 inches
plt.plot(x, y, marker='o', linestyle='-')
# Add annotations
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')
# textcoords :Specifies the coordinate system for the annotation text. "offset points" means the text is offset from the point.
# xytext: Specifies the offset for the text in points (10 points above the data point).
# ha :Horizontal alignment of the text.
# Add title and labels
plt. title('Line Chart with Annotations')
plt. xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
# Display grid
plt.grid (True)
# Show the plot
plt. show()
```

# arrow props

```python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# Create a line chart
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')
# Add a basic annotation with an arrow
plt.annotate( 'Annotation with arrow',
            xy=(3, 6),
            xytext=(4, 7),color='blue',
            arrowprops=dict(color='red', arrowstyle='->')
)
# Add title and labels
plt.title('Basic Annotation with Arrow')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
# Display grid
plt.grid(True)
# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# Create a line chart
plt.figure (figsize=(8, 6))
plt. plot(x, y, marker='o', linestyle='-')
# Add multiple annotations with different styles
annotations = [
              {'xy': (1, 2), 'text': 'Point 1', 'xytext': (2, 3), 'arrowprops': dict(color='blue', arrowstyle='->')},
              {'xy': (3, 6),'text': 'Point 2','xytext': (4, 5), 'arrowprops': dict(color='green', arrowstyle='->')},
              {'xy': (5, 10),'text': 'Point 3', 'xytext': (4, 11), 'arrowprops': dict(color='red', arrowstyle='->')}
]
for ann in annotations:
    plt.annotate(
            ann['text'],
          xy=ann['xy'],
          xytext=ann['xytext'],
          arrowprops=ann['arrowprops']
    )
plt.title('Multiple Annotations with Different Styles')
plt.xlabel('X-axis Label')
plt.ylabel ('Y-axis Label')
plt.grid(True)
plt.show()
```

# Radially displace pie chart wedge

```python
import matplotlib.pyplot as plt
#the slices are ordered and
# plotted counter-clockwise:
continents=['Asia','Europe','North America','South America','Australia','Africa','Antarctica']
area=[25,20,15,10,15,10,5]
explode=(0.1,0,0.1,0,0.1,0.1,0.1)
plt.pie(area,explode=explode,
        labels=continents,autopct='%1.1f%%',
        shadow=True,startangle=0,
        wedgeprops={'edgecolor': 'black', 'linewidth':2, 'antialiased': True})
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
#the slices are ordered and
#plotted counter -clockwise:
product='Product A','Product B','Product C','Product D'
stock=[20,30,25,20]
explode=(0.1,0,0.1,0)
plt.pie(stock,explode=explode,
        labels=product,autopct='%1.1f%%',
        shadow=True,startangle=90,
        wedgeprops={'edgecolor': 'black', 'linewidth': 2,
                    'antialiased': True})
plt.axis('equal')
plt.show()
```

# Box Plot

A box plot (also known as a box-and-whisker plot) is a graphical representation used to show the distribution of a dataset based on five key summary statistics, It provides a clear and concise way to visualize the spread, center, and variability of the data, as well as identify any potential outliers.


```python
import matplotlib.pyplot as plt
#  sample data
data = [20,15,18,22,24,19,30,25,34,28,25,42]
#create boxplot
plt.boxplot(data)
#Add title and labels
plt.title('Box Plot Example')
plt.xlabel('Data Category')
plt.ylabel('Value')
#show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
#  Data for Class A and Class B scores
class_A_scores = [55,60,62,65,67,70,72,75,50,85,90,95,98]
class_B_scores = [45,50,52,55,60,62,65,68,70,72,75,78,80]
#Plotting the boxplot
plt.figure(figsize=(8,6))
plt.boxplot([class_A_scores,class_B_scores],labels=['Class A','Class B'])
plt.title('Class Scores Distribution')
plt.xlabel('Classes')
plt.ylabel('Scores')
plt.show()
```

```python
import matplotlib.pyplot as plt
#  sample data
data = [20,15,18,22,24,19,30,32,100,28,25]
#create boxplot with color
plt.boxplot(data,patch_artist=True,boxprops=dict(facecolor='lightblue',color='blue'),
            whiskerprops=dict(color='blue'),capprops=dict(color='red'),
            flierprops=dict(markerfacecolor='green',marker='o',markersize=7))
#Add title and labels
plt.title('Box Plot Example')
plt.xlabel('Data Category')
plt.ylabel('Value')
# show plot
plt.show()
```

# Violinplot

```python
import matplotlib.pyplot as plt
import numpy as np
#Generatr random data for two different  categories
data1=np.random.normal(0,1,100)#100points,mean=0,std=1
data2=np.random.normal(5,1.5,100)#100points,mean=5,std=1.5

#combine the data into a list
data=[data1,data2]
#create the violinplot
plt.violinplot(data)
#add title and labels
plt.title('Simple Violin plot')
plt.xlabel(' Category')
plt.ylabel('Value')
#show plot
plt.show()
```

# Stem plot

```python
import matplotlib.pyplot as plt
import numpy as np
# Generate sample data
x=np.linspace(0,2*np.pi,10)
y=np.sin(x)
# Create the stem plot
plt.stem(x,y)
# Add title and labels
plt.title('Stem Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Generate sample time series data
dates=pd.date_range(start='2024-01-01',periods=10,freq='D')
values=np.random.randn(10) #random values for the time series

#Convert dates to numeric values(required for stem plot)
x=np.arange(len(dates))

# Create the stem plot
plt.stem(x,values,linefmt='g--',markerfmt='bo',basefmt='r-')

# Customize the X-axis with the actual date labels
plt.xticks(x,[date.strftime('%Y-%m-%d') for date in dates],rotation=45)


# Add title and labels
plt.title('Time Series Stem Plot')
plt.xlabel('Date')
plt.ylabel('Value')



# Show the plot
plt.show()
```
